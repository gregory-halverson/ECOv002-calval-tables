# Example: Plot PT-JPL-SM Model Results with quick_look_plot_single_model

This notebook demonstrates how to use the `quick_look_plot_single_model` function to visualize PT-JPL-SM model results against flux tower data.

```python
# Import required libraries and function
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from ECOv002_calval_tables import load_combined_eco_flux_ec_filtered
from ECOv002_calval_tables.plot_single_model import quick_look_plot_single_model
```

```python
big_df_ss = load_combined_eco_flux_ec_filtered()
big_df_ss
```

```python
# Call quick_look_plot_single_model for PT-JPL-SM and display plot
fig = quick_look_plot_single_model(
    big_df_ss=big_df_ss,
    time='example',
    model_col='PTJPLSMinst',
    model_name='PT-JPL-SM'
)

plt.show(fig)
```
